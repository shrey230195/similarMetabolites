```python
import pandas as pd
import numpy as np
import re

df = pd.read_excel('./uploads/mass_spec_data_assgnmnt.xlsx', sep='none')
# df.describe()
df.columns = [x.strip().replace(' ', '_') for x in df.columns]
df = df[df['Accepted_Compound_ID'].str.contains('_')==True]
PC = df[df['Accepted_Compound_ID'].str.endswith(' PC')==True]
LPC = df[df['Accepted_Compound_ID'].str.endswith('LPC')==True]
Plasmalogen = df[df['Accepted_Compound_ID'].str.endswith('plasmalogen')==True]

# value1 = df['Retention_time_(min)'].round()
# df.insert(2, 'Retention_time_round_off(min)', value1)

df1=df.drop(df.columns[[0, 1,2,3, 4]], axis=1)
# df1['avg'] = df1.mean(axis=1)
PC1=PC.drop(PC.columns[[0, 1,2,3, 4]], axis=1)
PC['mean']=PC1.mean(axis=1)
LPC1=LPC.drop(LPC.columns[[0, 1,2,3, 4]], axis=1)
LPC['mean']=LPC1.mean(axis=1)
Plasmalogen1=Plasmalogen.drop(Plasmalogen.columns[[0, 1,2,3, 4]], axis=1)
Plasmalogen['mean']=Plasmalogen1.mean(axis=1)

PC_index = np.arange(0, len(PC), 1)
PC['index']=PC_index
PC=PC.set_index('index')

print df['Accepted_Compound_ID']
```

```python
from scipy import stats

z_stat, p_val = stats.ranksums(PC['mean'],PC['mean'])  
print "MWW RankSum P for PC and LPC =", p_val 

z_stat, p_val = stats.ranksums(Plasmalogen['mean'],PC['mean'])  
print "MWW RankSum P for PC and Plasmalogen =", p_val 

z_stat, p_val = stats.ranksums(LPC['mean'],Plasmalogen['mean'])  
print "MWW RankSum P for LPC and Plasmalogen =", p_val 

f_val, p_val = stats.f_oneway(PC['mean'],LPC['mean'],Plasmalogen['mean'])  
  
print "One-way ANOVA P =", p_val  
```
